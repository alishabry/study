```python
df = pd.read_csv('titanic.csv')
```

```python
# 1.Выведите распределение переменной __Pclass__ для мужчин и женщин.

#     Сколько мужчин было в первом классе?
#     Сколько женщин было в третьем классе?

print(pd.crosstab(df['Pclass'], df['Sex']))
((df['Pclass'] == 1) & (df['Sex'] == 'male')).sum(), ((df['Pclass'] == 3) & (df['Sex'] == 'female')).sum()

```

```python
# 2.Создайте новый признак __AgeGroup__, который равен:

#     1, если пассажир был моложе 30 лет;
#     2, если пассажиру было от 30 до 60 лет;
#     3, если пассажир был старше 60 лет.
df['Age'].fillna(df['Age'].median())
df['AgeGroup'] = df['Age'].apply(lambda x: 1 if x < 30 else (2 if x <= 60 else 3))
df.head()
```

```python
#3. Выведите данные о всех пассажирах, которые сели в Cherbourg (Embarked='C') и были старше 50 лет.
df[(df['Embarked'] == 'C') & (df['Age'] > 50)]
```

```python
# 4.Выведите данные о всех женщинах младше 20 и старше 50 лет.
df[((df['Age'] < 20) | (df['Age'] > 50)) & (df['Sex'] == 'female')].head(10)
```

```python
# 5.Сколько пассажиров старше 70 лет было на борту?
(df['Age'] > 70).sum()
```

```python
# 6.Посчитайте средний возраст среди мужчин и среди женщин.
df.groupby('Sex')['Age'].mean()
```

```python
# 7.Посчитайте доли выживших в различных возрастных группах (новый признак **AgeGroup**).
df.groupby('AgeGroup')['Survived'].mean()
```

```python
# 8.Найдите максимальную стоимость билета в каждом из ценовых классов
df.groupby('Pclass')['Fare'].max()
```
